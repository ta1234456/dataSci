### Data Source
    
ข้อมูลเกี่ยวกับการจองโรงแรม จาก [Kaggle](https://www.kaggle.com/datasets/muhammaddawood42/hotel-booking-cancelations/data)

#### รายละเอียดของข้อมูล

1. hotel: ประเภทของโรงแรม
    * Resort Hotel โรงแรมเพื่อการพักผ่อน
    * City Hotel
2. is_canceled: สถานะการยกเลิก 
    * 1 : ยกเลิก
    * 0 : ไม่ยกเลิก
3. lead_time
4. arrival_date_year: ปีที่มาถึง
5. arrival_date_month: เดือนที่มาถึง
6. arrival_date_week_number: สัปดาห์ที่มาถึง
7. arrival_date_day_of_month: วันที่มาถึง
8. stays_in_weekend_nights: จำนวนคืนที่พักในช่วงสุดสัปดาห์
9. stays_in_week_nights: จำนวนคืนที่พักในช่วงสัปดาห์
10. adults: จำนวนผู้ใหญ่
11. children: จำนวนเด็ก
12. babies: จำนวนทารก
13. meal: ประเภทอาหาร
14. country: ประเทศของลูกค้า
15. market_segment: กลุ่มตลาด
16. distribution_channel: ช่องทางการจัดจำหน่าย
17. is_repeated_guest: ลูกค้าที่เคยมาเยือนอีกครั้ง (ใช่/ไม่ใช่)
18. previous_cancellations: จำนวนการยกเลิกก่อนหน้า
19. previous_bookings_not_canceled: จำนวนการจองที่ไม่ถูกยกเลิกก่อนหน้า
20. reserved_room_type: ประเภทห้องที่จอง
21. assigned_room_type: ประเภทห้องที่กำหนด
22. booking_changes: จำนวนการเปลี่ยนแปลงการจอง
23. deposit_type: ประเภทเงินมัดจำ
24. agent: ตัวแทนการจอง
25. company: บริษัทที่ทำการจอง
26. days_in_waiting_list: จำนวนวันที่อยู่ในรายการรอ
27. customer_type: ประเภทลูกค้า
28. adr: ราคาต่อคืนเฉลี่ย
29. required_car_parking_spaces: จำนวนที่จอดรถที่ต้องการ
30. total_of_special_requests: จำนวนคำขอพิเศษ
31. reservation_status: สถานะการจอง
32. reservation_status_date: วันที่ของสถานะการจอง
33. name: ชื่อลูกค้า
34. email: อีเมลลูกค้า
35. phone-number: หมายเลขโทรศัพท์ลูกค้า
36. credit_card: ข้อมูลบัตรเครดิต


ชุดข้อมูลนี้มีข้อมูลเชิงลึกเกี่ยวกับการจองโรงแรม รวมถึงสถานะการยกเลิก โดยประกอบด้วยตัวแปรต่างๆ เช่น วันที่จอง สถานะการยกเลิก เวลานำหน้า ประเภทลูกค้า และประเภทโรงแรม (ในเมืองหรือรีสอร์ท) เหมาะสำหรับการวิเคราะห์แนวโน้มการยกเลิกการจอง เข้าใจพฤติกรรมลูกค้า และคาดการณ์ความน่าจะเป็นในการยกเลิก ชุดข้อมูลนี้เหมาะสำหรับโปรเจกต์ด้านวิทยาการข้อมูลที่เกี่ยวข้องกับการจำแนกประเภท การวิเคราะห์ข้อมูลตามเวลา หรือการสร้างโมเดลคาดการณ์เพื่อช่วยลดการยกเลิกและเพิ่มประสิทธิภาพกลยุทธ์การจอง.
